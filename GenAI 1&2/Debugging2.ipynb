```python
import faiss
import numpy as np
import google.generativeai as genai

# Configure Gemini API Key
genai.configure(api_key="AIzaSyBcJ1z5FQPOgE3-sY0w9O1xQJQhrYC4AvE")

# FAISS Index with correct embedding size
d = 768  # Gemini embedding model returns 768-d vectors
index = faiss.IndexFlatL2(d)

# Function to get embeddings using Gemini's embedding API
def get_embedding(text):
    response = genai.embed_content(model="models/embedding-001", content=text, task_type="retrieval_document")
    return np.array(response['embedding'], dtype='float32').reshape(1, -1)  # Correct API call

# Documents to be indexed
documents = ["AI is transforming industries", "FAISS is great for vector search", "Generative models are powerful"]

# Generate embeddings and add them to FAISS index
doc_embeddings = np.vstack([get_embedding(doc) for doc in documents])  # Correct stacking
index.add(doc_embeddings)

# Query embedding
query = "How is AI transforming industries?"
query_vector = get_embedding(query)  

# Search for the most relevant document
_, indices = index.search(query_vector, k=1)
retrieved_doc = documents[indices[0][0]]

print("Retrieved Document:", retrieved_doc)
```
